# Testing server function with image input and output

```python
import requests
import cv2
import base64
import numpy as np
from google.colab.patches import cv2_imshow

# function to convert a string to an image as a np.array
def img_from_string(img_str):
  orig = base64.b64decode(img_str.encode("utf-8"))
  np_img = np.frombuffer(orig, dtype=np.uint8)
  return cv2.imdecode(np_img, cv2.IMREAD_COLOR)
```

* Replace `YOUR_URL` with your endpoint (e.g. https://github.com/author)
* Upload an image to colab
* Set `image_name` to the name of the image that was uploaded  

```python
url1 = "YOUR_URL"
image_name = ""
```

Send a `POST` request to Google Cloud Functions. The status of the Response should be 200.\
**NOTE:** all the cloud functions are deployed as `POST` methods.

```python
r = requests.post(url1, files=dict(img=open(image_name, "rb")))
print(r)
```

```python
result = img_from_string(r.json()["content"])
print(result.shape)
```

```python
cv2_imshow(result)
```
